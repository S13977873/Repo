```python
import pandas as pd
import plotly
import plotly.graph_objects as go
import plotly.express as px
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
# import plotly.offline as pi
from matplotlib.colors import LinearSegmentedColormap
from plotly.subplots import make_subplots
from scipy.stats import linregress
import geopandas as gpd
import numpy as np
plotly.offline.init_notebook_mode(connected=False)
```

```python
# Read the dataset into a DataFrame
df = pd.read_csv("merged_df_happy_2019.csv")

# Read the shapefile containing the world map
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Create a dictionary that maps the country names in your dataset to the corresponding country names in the world GeoDataFrame
name_map = {
    'United States': 'United States of America',
}

# Replace the country names in your DataFrame using the name_map dictionary
df['Country or region'] = df['Country or region'].replace(name_map)

# Merge the two DataFrames on the country name
merged = world.merge(df, left_on='name', right_on='Country or region', how='left')

# Create a custom color map
cmap = mcolors.LinearSegmentedColormap.from_list("mycmap", [[0, '#0000ff'], [0.25, '#9400d3'], [0.5, '#ff69b4'], [0.75, '#ffa500'], [1, '#ffff00']])

# Plot the world map with the happiness score as the color
fig, ax = plt.subplots(1, 1, figsize=(30, 10))
merged.plot(column='happiness_score', ax=ax, cmap=cmap, missing_kwds={'color': 'lightgrey'})

# Create a custom colorbar
norm = mcolors.Normalize(vmin=merged['happiness_score'].min(), vmax=merged['happiness_score'].max())
sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])
cbar = fig.colorbar(sm)
cbar.set_ticks([merged['happiness_score'].min(), merged['happiness_score'].max()])
cbar.set_ticklabels(['Min-Happiness', 'Max-Happiness'], fontsize=30, fontname='Comic Sans MS', color='#6d00df')  

# Remove x and y axis tick labels
ax.set_xticklabels([])
ax.set_yticklabels([])

# Add a title to the plot with pink color
ax.set_title('Het gevoel van gelukzaligheid op de aarde', fontsize=45, color='#6d00df')

# Remove the frame around the colorbar
cbar.outline.set_visible(False)

# Remove the newspaper watermark
ax.set_axis_off()

plt.show()

{
    "tags": [
        "hide-input",
    ]
}
```

```python
# Lees het CSV-bestand met gegevens over landen
df = pd.read_csv("merged_df_happy_2019.csv")

# Bereken de gemiddelde geluksscore, het gemiddelde overheidsvertrouwen en de gemiddelde vrijheid
mean_happiness = df['happiness_score'].mean()
mean_trust = df['government_trust'].mean()
mean_freedom = df['freedom'].mean()

# Centreer de gegevens rond het nulpunt
df['happiness_score_centered'] = df['happiness_score'] - mean_happiness
df['government_trust_centered'] = df['government_trust'] - mean_trust
df['freedom_centered'] = df['freedom'] - mean_freedom
# Create a custom colormap that goes from blue to purple to pink to orange to yellow
cmap = LinearSegmentedColormap.from_list('BluePurplePinkOrangeYellow', ['#0000ff', '#9400d3', '#ff69b4', '#ffa500', '#ffff00'])
norm = plt.Normalize(vmin=df['happiness_score'].min() - 1, vmax=df['happiness_score'].max() + 1)
df['happiness_color'] = df['happiness_score'].apply(lambda x: cmap(norm(x)))

# Convert the RGBA values to a valid CSS color string
df['happiness_color'] = df['happiness_color'].apply(lambda x: mcolors.to_hex(x))

# Create a figure with a custom size and one subplot
fig = make_subplots(rows=1, cols=1)

# Plot the data as text instead of points with a larger text size and custom font
for i, row in df.iterrows():
    fig.add_trace(go.Scatter(
        x=[row['government_trust_centered']],
        y=[row['freedom_centered']],
        mode='text',
        text=[row['Country'][:3]],
        hovertext=[row['Country']],
        textposition='bottom center',
        textfont=dict(
            size=14,
            color=row['happiness_color'],
            family='Arial Black'
        ),
        hovertemplate='<b>%{hovertext}</b>',
        hoverlabel=dict(
            bgcolor='white',
            font=dict(
                size=16,
                color=row['happiness_color']
            )
        )
    ))

# Calculate the linear regression and add a correlation line
slope, intercept, r_value, p_value, std_err = linregress(df['government_trust_centered'], df['freedom_centered'])
x_range = [df['government_trust_centered'].min(), df['government_trust_centered'].max()]
y_range = [slope * x + intercept for x in x_range]
fig.add_trace(go.Scatter(
    x=x_range,
    y=y_range,
    mode='lines',
    line=dict(
        color='black',
        width=2
    ),
    showlegend=False
))

# Configure the layout of the plot
fig.update_layout(
    title='Verband tussen overheidsvertrouwen en vrijheid',
    title_font=dict(
        size=24,
        color='#6d00df'
    ),
    showlegend=False,
    xaxis=dict(
        title='Overheidsvertrouwen',
        zeroline=True,
        zerolinecolor='black',
        showline=False,
        showgrid=False,
        linecolor='black',
        mirror=True,
        showticklabels=True
    ),
    yaxis=dict(
        title='Vrijheid',
        zeroline=True,
        zerolinecolor='black',
        showline=False,
        showgrid=False,
        linecolor='black',
        mirror=True,
        showticklabels=False
    ),
    plot_bgcolor='white',
    paper_bgcolor='white',
    width=800,
    height=500
)

# Create a colorbar that matches the colors used in the plot
fig.add_trace(go.Scatter(
    x=[df['government_trust_centered'].max() + 0.2],
    y=[df['freedom_centered'].max()],
    mode='markers',
    marker=dict(
        colorscale=[(0.0, '#0000ff'), (0.25, '#9400d3'), (0.5, '#ff69b4'), (0.75, '#ffa500'), (1.0, '#ffff00')],
        colorbar=dict(
            title='Geluksscores'
        ),
        cmin=df['happiness_score'].min(),
        cmax=df['happiness_score'].max(),
        showscale=True,
        opacity=0
    ),
    showlegend=False
))

# Show the interactive plot
fig.show()
{
    "tags": [
        "hide-input",
    ]
}
```

Uit onze data blijkt dat Landen die een hoge gemiddelde temperatuur ervaren over het algemeen minder gelukkig zijn. In grafiek 5 is deze negatieve correlatie te zien. Daarnaast is in de figuur ook te zien dat alle landen in onze dataset die en zeer lage geluksscore hebben ook een zeer hoge gemiddelde temperatuur ervaren. Dit zou kunnen zijn omdat deze landen langdurig hoge temperaturen ervaren en dit soort perioden een negatief effect heeft op gelukkigheid (Rehdanz et al.,). Ook interessant is dat koudere landen geen vergelijkbare geluksscores hebben in onze data door erg koude perioden van tijd, aangezien deze ook een negatief effect hebben op de gelukkigheid van mensen (Rehdanz et al.,).

```python
merged_streamlined_df = pd.read_csv("Weerdata.csv")

x1 = merged_streamlined_df["happiness_score"]
y1 = merged_streamlined_df["AverageTemperature"]

# Fit linear regression model
k1, m1 = np.polyfit(x1, y1, 1)

# Create a color scale
colorscale = [[0.0, 'darkblue'], [0.25, 'purple'], [0.5, '#ff69b4'], [0.75, 'orange'], [1.0, 'yellow']]

# Create the first scatter plot
fig = go.Figure()
fig.add_trace(go.Scatter(x=x1, y=y1, mode='markers', marker=dict(size=10, color=x1, colorscale=colorscale, showscale=True, colorbar=dict(title='Geluksscore'))))
fig.add_trace(go.Scatter(x=x1, y=k1*x1 + m1, mode='lines', name='Regression line', line=dict(color='black')))

fig.update_layout(
    title= 'Correlatie tussen de geluksscores en gemiddelde temperatuur per land',
    height=600,
    xaxis_title='Geluksscore',
    yaxis_title='Gemiddelde Temperatuur',
    showlegend=False
)

fig.show()
{
    "tags": [
        "hide-input",
    ]
}
```
